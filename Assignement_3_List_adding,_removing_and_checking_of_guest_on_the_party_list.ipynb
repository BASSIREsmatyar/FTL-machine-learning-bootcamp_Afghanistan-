# Assignement #3: List; adding, removing and checking of guest on the party list

```python
guest_list = []  # Create an empty guest list

# define three functions for each one (add, remove, and checking guest)

def add_guest(*names): #*names for adding one and more at once
    guest_list.extend(names)
    print(f"{names} has been added to the list.")

def remove_guest(names):
    if names in guest_list:
        guest_list.remove(names)
        print(f"{names} has been removed from the list.")
    else:
        print(f"{names} is not on the list.")

def check_guests(names):
    if names in guest_list:
        print(f"{names} is on the guest list.")
    else:
        print(f"{names} is not on the list.")
        if not guest_list:  # Check if the list is empty
            print("The guest list is empty.")
# Example
add_guest( "Hamid", "Bassir", "Ali", "Ahmet")
remove_guest("Ali")
remove_guest("Zaman")
check_guests("Bassir")
check_guests("Mohammad")
print(guest_list)
```
