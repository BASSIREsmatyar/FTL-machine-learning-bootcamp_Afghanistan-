# Assignement #4: using Python dictionaries and tuples by building a guest management system for a party.

```python
# create empty dict
guests = {}

#adding guests

guests ["Alice"]= ( 28, "alice@email.com")
guests ["Bob"]= ( 35, "bob@email.com")
guests ["Charlie"]= ( 30, "charlie@email.com")

print("The first list is", guests)

# Updating the guest list
guests ["David"] = (22, "david@email.com")  # added David

print( "\nThe second list is", guests)

del guests ["Bob"] # Bob has been deleted

print ("\nThe third list is", guests)

# Guest Information Retriever

def get_guest_info(guest_name):
    if guest_name in guests:
        age, email = guests[guest_name]
        print(f"\n{guest_name} ({age}) is coming to the party! Email: {email}.")
    else:
        print(f"\n{guest_name} is not on the list.")

# Example

get_guest_info("Alice")
get_guest_info("Bob")
get_guest_info("David")

# Guest Counter

guest_count = len(guests)
print("\nThere are", guest_count, "guests invited.")
```
