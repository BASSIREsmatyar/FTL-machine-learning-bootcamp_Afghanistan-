```python
import math
```

```python
class circle:
  def __init__(self,r):
    self.r=r
  def area(self):
    return math.pi*self.r**2
  def perimeter(self):
    return 2*math.pi*self.r
  circle = circle(5)
  print(circle.area())
  print(circle.perimeter())
```

```python
class calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    return self.a+self.b
  def subtract(self):
    return self.a-self.b
  def multiply(self):
    return self.a*self.b
  def divide(self):
    return self.a/self.b
  calculator = calculator(10,6)
  print(calculator.add())
  print(calculator.subtract())
  print(calculator.multiply())
  print(calculator.divide())
```

```python
class personal_information:
  def __init__(self,name,country,date_birth):
    self.name=name
    self.country=country
    self.date_birth=(date_birth)
  def age(self):
    return 2024-self.date_birth
  def name_func(self):
    return self.name
  def country_func(self):
    return self.country

person = personal_information( "Bassir", "Afghanistan", 1996)
print(person.age())
print(person.name_func())
print(person.country_func())
```
